```python
# center limit theorem (CLT)
```

```python
import random
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
random.seed(1)
mean=[np.mean(np.random.randint(1,100,50)) for i in range(1000)]
sns.distplot(mean,label="gaussian distribution")
plt.title("CLT")
plt.grid()
plt.legend()
plt.show()
```
